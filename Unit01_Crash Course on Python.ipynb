### 01 - 奇偶數辨別（Odd or even）
#### 當使用者輸入一個的數字，它能夠辨別奇偶，並輸出檢驗結果給使用者。

> 題目：輸入一數字判斷奇數或偶數<br>
輸入(Input): 一個數字 <br>
輸出(Output):奇數或偶數

* 進階: 可以限制輸入一個介於一定範圍（例如 1 到 1000）的數字

````python
def evenodd(input):
    if 1 <= input <= 1000:
        if input%2==0:
            return "even"
        return "odd"
    else:
        return "out of range, please input any integer between 1 and 1000"

inputnum = int(input('Please key in input number: '))
EvenOddOutput = evenodd(inputnum)
print('Input number '+ str(inputnum) + ' is ' + EvenOddOutput)
````

> 題目：輸入一數字判斷奇數或偶數<br>

輸出(Output): 模板故事

* 進階: 可以讓程式重複執行

````python
from colorama import Fore, Back, Style
#6.5 creedmoor DOPE dictionary
dope_evelavtion = {0:0, 100:0, 200:0.3, 300:0.9, 400:1.8, 500:2.7, 600:3.8, 700:5, 800:6.3, 900:7.8, 1000:9.5}
dope_windage10mph = {0:0, 100:0.2, 200:0.3, 300:0.3, 400:0.7, 500:0.9, 600:1.1, 700:1.3, 800:1.5, 900:1.8, 1000:2.1}

#Target is up!
print('')
print('Spotter: TargetIsUp!')
print('')
TSector = input('WhatSector? ') 
TargetDescription = input('Target? ') 
#Ranging
mRAD = input('mil? ') 
TargetSizecm = input('size in cm? ') 

print('')
print(Fore.BLUE + 'Spotter: TargetIsUp! Sector '+ TSector +', '+ TargetDescription + '.')
print(Fore.BLUE + 'Shooter: Roger that! Postive ID, Sector '+ TSector +', ' + TargetDescription + ', on sight.')
TargetRange=round(float(TargetSizecm)/100/float(mRAD)*1000,0)
print(Fore.BLUE + 'Shooter: Target size ' + str(TargetSizecm) + ' cm, ' + str(mRAD) + ' mils in scope, distance ' + str(TargetRange) + ' meters.')
print(Fore.BLUE + 'Spotter: Confirm distance ' + str(TargetRange) + ' meters.')
print(Fore.BLACK)

PositiveHit = ''
engageattemps = 0
while PositiveHit!='yes' and engageattemps<3:
    engageattemps += 1
    if engageattemps < 3:   
        #DOPE
        winddir = input('wind direction?' )
        wspeed = input('wind speed?' )
        
        ditindex = int(TargetRange)//100*100        
        ditindex1 = ditindex+100
        Ratio = (TargetRange-float(ditindex))/100
        
        #read dope dictionary, do interpolation
        elevcorr = round(dope_evelavtion[ditindex] + Ratio*(dope_evelavtion[ditindex1]-dope_evelavtion[ditindex]),1)
        windcorr = dope_windage10mph[ditindex] + Ratio*(dope_windage10mph[ditindex1]-dope_windage10mph[ditindex])
        windcorr = round(float(wspeed)/10,1) #find wind correction according to wind speed  
        
        #ReportAndCorrect
        PositiveHit = input('hit?')
        ShotLocation = input('ShotLocation?')
        print('')
        print(Fore.BLUE + 'Spotter: ' + str(elevcorr) + ' mils up, ' + str(windcorr )+ ' mils to the ' + winddir +'.')
        print(Fore.BLUE + 'Shooter: Shooter Ready.')
        print(Fore.BLUE + 'Spotter: Spotter ready. Send it.')
        if PositiveHit == 'yes':
            print(Fore.BLUE + 'Spotter: Impact! Center of mass, good hit, good hit.')
        else: 
            #report shot location for correction
            print(Fore.BLUE + 'Spotter: ' + ShotLocation)
    else:
         #too many attempts, break contact
         print(Fore.BLUE + 'Spotter: Let\'s get the hell out of here!')
    print('')
````

````python
# Get user email address
email = input("What is your email address?: ").strip()

# Slice out the user name
user_name = email[:email.index('@')]

# Slice the domain name
domain_name = email[email.index('@')+1:len(email)]

# Format message
output = "Your username is '{}' and your domain name is '{}'".format(user_name,domain_name)

# Display output message
print(output)


domain = {'gmail.com':'Google','yahoo.com.tw':'Yahoo','ntu.edu.tw':'臺大','hotmail.com.tw':'Hotmail','hotmail.com':'Hotmail'}
if domain_name in domain:
    output = "這是註冊在 {} 之下的 Email 地址'".format(domain[domain_name])
else:
    
    output = "這是在 {} 之下自定義域".format(domain[domain_name])
print(output)
````

### 09 - 回文判斷（Is a palindrome）
#### 請使用者輸入單字，判斷它是否為回文，也就是該單字前後對稱，例如 madam，從前讀到後或是從後讀到前的順序都是 madam。

> 題目：<br>
輸入(Input): 雨 滋 春 樹 碧 連 天  天 連 碧 樹 春 滋 雨 <br>
輸出(Output):The text you have entered is a palindrome!<br>
輸入(Input): 資工訓練班 <br>
輸出(Output):The text you have entered is not a palindrome.


````python
def convertInputString():
    #請使用者輸入一串字串，先轉換字串為統一小寫
    rawInput = input("\nPlease enter a word, phrase, or a sentence \nto check if it is a palindrome: ") 
    rawString = rawInput.lower()
    rawList = list(rawString)
    return rawList

def stripAnalphabetics(dirtyList): 
    analphabeticList = [" ", "-", ".", ",", ":", ";", "!", "?", "'", "\""] 
    #去除相關標點符號
    for character in analphabeticList:
        if character in dirtyList:
            dirtyList.remove(character)
            return stripAnalphabetics(dirtyList)
    return dirtyList

def runPalindromeCheck(straightList):
    reversedList = straightList[::-1]  # ::-1 reverse the list
    if straightList == reversedList:
        return "The text you have entered is a palindrome!" 
    else: 
        return "The text you have entered is not a palindrome." 

def main(): 
    print("\nPalindrome checker") 
    originalList = convertInputString()
    originalList = stripAnalphabetics(originalList)
    palindromeCheck = runPalindromeCheck(originalList)
    print(palindromeCheck)

main() 
````
